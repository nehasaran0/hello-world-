```
import matplotlib.pyplot as plt
x=[1,2,14]
y=[4,6,9]
x1=[7, 6,8]
y1=[6,15,7]
plt.plot(x,y, label ="line")
plt.plot(x1,y1) 
plt.title("This is a line")
plt.legend()
plt.grid(True, color ='red') 
plt.show()
```

```
plt.plot([1,2,3,4], [5,6,4,9], 'r*', label ="line1")
plt.plot([4,7,4,8],[2,4,7,6], "ro-", label ="line2")
plt.legend()
plt.grid()
plt.show()
```
